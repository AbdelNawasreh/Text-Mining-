```python
import nltk
```

```python
from nltk.corpus import brown
```

```python
borwn = brown.tagged_words()
word_fd = nltk.FreqDist(borwn) 
[w[0] for (w, _) in word_fd.most_common() if w[1] == 'NN'] 
```

```python
borwn = brown.tagged_words()
word_fd = nltk.FreqDist(borwn) 
[w[0] for (w, _) in word_fd.most_common() if w[1] == 'NNS'] 
```

```python
from nltk.corpus import brown
brown_tagged = brown.tagged_words()
tag_fd = nltk.FreqDist(tag for (word, tag) in brown_tagged)
tag_fd.most_common()
```

```python
tag_fd.most_common(20)
```
